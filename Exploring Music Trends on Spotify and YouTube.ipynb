## Introduction

Spotify and YouTube are two of the most popular streaming platforms for music worldwide. While Spotify is primarily a music streaming service that allows users to listen to millions of songs on demand, YouTube is a video sharing platform that also offers music videos, live performances, and other music-related content.

## Data Cleaning

```python
# Remove unused columns
df = df.drop(['Unnamed: 0','Uri','Url_spotify', 'Url_youtube', 'Description'], axis=1)
```

```python
# Check null values
df.isna().sum()
```

```python
# Remove null values
df.dropna(inplace=True)
```

```python
# Check duplicates
duplicates = df.duplicated(subset='Track', keep=False)
duplicates.value_counts()
```

There are 4668 duplicate values in this dataset.

```python
# Delete duplicates
df= df.drop_duplicates(subset='Track')
```

## Exploratory Data Analysis

```python
# Set default values for the graph plot
pio.templates['plotly'].layout.xaxis.title.font.size = 18
pio.templates['plotly'].layout.yaxis.title.font.size = 18
pio.templates['plotly'].layout.title.font.size = 24
pio.templates['plotly'].layout.font.family = 'Arial'
pio.templates['plotly'].layout.colorway=['#057fdc']
```

```python
album_types = df.groupby('Album_type', as_index=False)['Album'].count()

fig = px.pie(album_types, values='Album', names='Album_type',color_discrete_sequence=[ '#057fdc','green', 'orange'])
fig.update_traces(textinfo='percent+label', textfont_size=18, marker=dict(line=dict(color='white', width=1.5)))
fig.update_layout(title = 'Type of Albums', title_x =0.5, width=1000, height=600)
img_bytes = pio.to_image(fig, format="png", width=1000, height=600)
Image(img_bytes)
```

The pie chart represents the distribution of different types of music releases, with albums contributing to the majority of the releases at 74.4%. Singles contribute to 22.1% of the releases, and compilations make up the remaining 3.5%. This indicates that albums are the most popular form of music release, followed by singles and then compilations.

#### Top Ten Songs and Artists

```python
views = df.groupby(['Views','Stream'], as_index=False)['Track'].sum().sort_values('Views', ascending= False)[:10]

views['Views (millions)'] = views['Views'] / 1000000
views['Stream (millions)'] = views['Stream'] / 1000000
stream = views.sort_values('Stream', ascending=False)[:10]
```

```python
# Create subplot figure with two horizontal subplots
fig = make_subplots(rows=1, cols=2)

# Add bar plots to subplots
fig.add_trace(px.bar(views, x='Views', y='Track', orientation='h',color_discrete_sequence=['#057fdc']).data[0], row=1, col=1)
fig.update_xaxes(title='Views (billions) on YouTube',row=1, col=1)
fig.update_yaxes(categoryorder='total ascending', row=1, col=1)

bar2 =fig.add_trace(px.bar(stream, x='Stream', y='Track', orientation='h',color_discrete_sequence=['#057fdc']).data[0],row=1, col=2)
fig.update_xaxes(title='Stream (billions) on Spotify', row=1, col=2)
fig.update_yaxes(categoryorder='total ascending', row=1, col=2)

fig.update_layout(title = 'Top Ten Songs By Most Views on YouTube and Stream on Spotify', title_x=0.5,
                  showlegend=False, height= 600)
img_bytes = pio.to_image(fig, format="png", width=1000, height=600)
Image(img_bytes)
```

The reason why songs with the most views on YouTube are not the top songs on Spotify could be due to a variety of factors. Some songs that are popular on YouTube may not be as popular on Spotify because the audiences of these two platforms have different preferences. Additionally, the algorithms used by each platform to recommend and promote songs may also differ, leading to differences in the popularity of songs on each platform. Finally, other factors such as marketing and promotion by artists and record labels may also play a role in the popularity of songs on each platform.

```python
audio_features = df[['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo']]

# Create a correlation matrix
corr_matrix = audio_features.corr()

fig = px.imshow(corr_matrix, color_continuous_scale='blues')
fig.update_layout(title='Heatmap of Audio Feature Correlations',title_x=0.5, title_xanchor ='center',width = 1000, height = 600 )
for i in range(corr_matrix.shape[0]):
    for j in range(corr_matrix.shape[1]):
        fig.add_annotation(
            x=j,
            y=i,
            text=f"{corr_matrix.iloc[i, j]:.2f}",
            showarrow=False,
            font=dict(color="black" if np.abs(corr_matrix.iloc[i, j]) < 0.9 else "white")
        )
img_bytes = pio.to_image(fig, format="png", width=1000, height=600)
Image(img_bytes)
```

This heatmap represents the correlation between the audio features. We can see there are some positive and negative correlation between audio features

```python
# Correlation coefficient
corr = np.corrcoef(df['Loudness'], df['Energy'])[0,1]

fig = px.scatter(df, x='Loudness', y='Energy')
fig.update_layout(title = 'Loudness vs Energy', title_x = 0.5, title_xanchor = 'center',width= 1000, height = 600)
fig.update_xaxes(title ='Loudness (db)')
fig.add_annotation(x=0,y=1.05, text=f"Correlation: {corr:.2f}", showarrow=False, xref="paper", yref="paper", font=dict(size=16))
img_bytes = pio.to_image(fig, format="png", width=1000, height=600)
Image(img_bytes)
```

Loudness and Energy are two audio features that are commonly used to describe music. Loudness refers to the perceived volume of a song, while Energy refers to the intensity or activity level of a song. In general, songs with high loudness tend to have high energy as well, as they are often more intense and dynamic. However, we can see there are  some songs that have high energy without being particularly loud, or vice versa. 

```python
# Correlation coefficient
corr2 = np.corrcoef(df['Acousticness'], df['Energy'])[0,1]
corr3 = np.corrcoef(df['Acousticness'], df['Loudness'])[0,1]

# Left graph
fig =make_subplots(rows=1, cols =2)
fig.add_trace(px.scatter(df, x='Acousticness', y='Energy').data[0], row=1, col=1)
fig.update_xaxes(title = 'Energy', row=1, col = 1)
fig.update_yaxes(title = 'Acousticnes', row =1, col =1)
fig.add_annotation(x=0,y=1.05, text=f"Correlation: {corr2:.2f}", showarrow=False, xref="paper", yref="paper", font=dict(size=15))

# Right graph
fig.add_trace(px.scatter(df, x='Acousticness', y='Loudness').data[0], row=1, col=2)
fig.update_yaxes(title = 'Loudness', row =1, col =2)
fig.update_xaxes(title = 'Acousticness', row =1, col =2)
fig.add_annotation(x=1,y=1.05, text=f"Correlation: {corr3:.2f}", showarrow=False, xref="paper", yref="paper", font=dict(size=15))

fig.update_layout(title = 'Acousticness vs Energy and Loudness', title_x = 0.5, title_xanchor = 'center',width= 1000, height = 600)

img_bytes = pio.to_image(fig, format="png", width=1000, height=600)
Image(img_bytes)
```

A track with high acousticness is likely to be composed of mostly acoustic instruments, such as guitar, piano, or strings, whereas a track with low acousticness is likely to be composed of mostly electronic instruments. A negative correlation between energy (left side) and acousticness indicates that as the energy of a track increases, the amount of acoustic instruments decreases, and vice versa. We can see the similar pattern for loudness (right side) also. Where acousticnes tend to decreases as loudness increase.


Electronic music tends to have more compression and limiting applied to it, which reduces the dynamic range and makes it sound louder. Acoustic music tends to have more natural dynamics and less processing, which makes it sound quieter. Also, electronic music often uses sounds that have more high-frequency content, which can make them seem louder than sounds that have more low-frequency content. Acoustic music often uses sounds that have more low-frequency content, which can make them seem quieter than sounds that have more high-frequency.

However, correlation doesn't mean causation. A strong correlation between two audio features means that they tend to vary together, but it does not necessarily mean that one feature causes the other to change.
